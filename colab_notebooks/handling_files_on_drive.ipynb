# Building the Self-Supervision dataset

## Setup

```python
import scipy.io
import numpy as np
from shutil import copyfile
from os.path import join
```

### Path

```python
%cd /content/drive/MyDrive/Università/Data Science/Second Year/Second Semester/Path of Excellence/Data
```

```python
source_folder = "./PRW-v16.04.20/"
destination_folder = "./PWR_self_supervised/"
```

### getting file names

```python
mat = scipy.io.loadmat(source_folder + "frame_train.mat")
```

```python
file_names = list(map(lambda name: name.item(), mat["img_index_train"].squeeze()))
```

train test split

```python
train_files, val_files = np.split(file_names, [round(len(file_names)*0.8)])
```

## Copyng the files

```python
for file_name in train_files:
    copyfile(join(source_folder, "frames", file_name+".jpg"), 
             join(destination_folder, "train", file_name+".jpg"))
```

```python
for file_name in val_files:
    copyfile(join(source_folder, "frames", file_name+".jpg"), 
             join(destination_folder, "val", file_name+".jpg"))
```
